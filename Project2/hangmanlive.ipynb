```python
import import_ipynb
from words import words_list
import random
```

```python
def getRandom():
    word=random.choice(words_list)
    return word.upper()

def play(word):
    word_completion="_"*len(word)
    guessed=False
    guessed_letters=[]
    guessed_words=[]
    tries=6
    
    print("Let's play Hangman!")
    print(displayHangman(tries))
    print(word_completion,"(the lenght of the word is",len(word),")")
    
    while not guessed and tries>0:
        guess=input("Enter your guess: ")
        guess=guess.upper()
        if len(guess)==1 and guess.isalpha():
            if guess in guessed_letters:
                print("You have already guessed the letter",guess)
            elif guess not in word:
                print(guess,"is not in the word")
                tries-=1
                guessed_letters.append(guess)
            else:
                print("Good job",guess,"is in the word")
                guessed_letters.append(guess)
                words_as_list=list(word_completion)
                indices=[i for i,letter in enumerate(word) if letter==guess]
                for index in indices:
                    words_as_list[index]=guess
                word_completion="".join(words_as_list)
                if "_" not in word_completion:
                    guessed=True
               
        elif len(guess)==len(word) and guess.isalpha():
            if guess in guessed_words:
                print("You have already guessed the word",guess)
            elif guess!=word:
                print(guess,"is not the word")
                tries-=1
                guessed_words.append(guess)
            else:
                guessed=True
                word_completion=word
        else:
            print("Invalid Input")
            
        print(displayHangman(tries))
        print(word_completion)
        
        if guessed and tries>0:
            print("Congrats, you guessed the word! You win!")
        if not tries>0:
            print("Sorry, you ran out tries. The word was "+ word+".")
            print("Better luck next time")
            
            
def displayHangman(tries):
    stages=[
        '''
            ------------
            |       |
            |       O
            |      \\|/ 
            |       |
            |      / \\
            _
        ''' ,
        '''
            ------------
            |       |
            |       O
            |      \\|/ 
            |       |
            |        \\
        
        ''' ,
        '''
            ------------
            |       |
            |       O
            |      \\|/ 
            |       |
            |       
        
        ''' ,
        '''
            ------------
            |       |
            |       O
            |      \\| 
            |       |
            |      
        
        ''' ,
        '''
            ------------
            |       |
            |       O
            |       |
            |       |
            |      
        
        ''' ,
        '''
            ------------
            |       |
            |       O
            |       
            |       
            |      
        
        ''' ,
        '''
            ------------
            |       |
            |       
            |       
            |       
            |      
        
        '''
    ]
    return stages[tries]
        
def main():
    word=getRandom().upper()
    play(word)
    
    while input("Do you want to play again(Y/N): ").upper()=="Y":
        word=getRandom()
        play(word)
    
main()
```
